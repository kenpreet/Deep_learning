```python
import pandas as pd
import numpy as np

data = pd.read_csv("multiple_linear_regression_dataset.csv")

print(data.head())
print(data.columns)
print(data.shape)

X = data[["age", "experience"]].values
y = data["income"].values

n_features = X.shape[1]
w = np.zeros(n_features)
b = 0.0

def predict(X, w, b):
    return X.dot(w) + b

def mean_squared_error(y, y_hat):
    return ((y_hat - y) ** 2).mean()

def compute_gradients(X, y, y_hat):
    N = len(y)
    error = y_hat - y
    dw = (2 / N) * X.T.dot(error)
    db = (2 / N) * error.sum()
    return dw, db

def update_parameters(w, b, dw, db, lr):
    w = w - lr * dw
    b = b - lr * db
    return w, b

lr = 0.00001
epochs = 2000

for epoch in range(epochs):
    y_hat = predict(X, w, b)
    loss = mean_squared_error(y, y_hat)
    dw, db = compute_gradients(X, y, y_hat)
    w, b = update_parameters(w, b, dw, db, lr)
    if epoch % 200 == 0:
        print("Epoch", epoch, "Loss:", loss)

print("Final Weights:", w)
print("Final Bias:", b)

new_person = np.array([35, 6])
predicted_income = new_person.dot(w) + b
print("Predicted Income:", predicted_income)
```

Step 2 Reflection

Shape of X: (N, 2)
Shape of y: (N,)

X has 2 columns because there are 2 input variables.
y has only one column because we are predicting one numeric value.

Step 3 Reflection

We need one weight per feature because each feature contributes differently to the prediction.
Bias is separate because it shifts the prediction independently of inputs.
Initializing with large values is risky because predictions may explode and gradients may become unstable.

Step 4 Reflection

There is no activation function because this is regression.
Predicted values can be any real number.
Unlike logistic regression, we are not converting outputs into probabilities.

Step 5 Reflection

We square the error to penalize large mistakes more strongly.
If one prediction is very wrong, its penalty becomes very large.
Absolute error is less sensitive to large deviations and is harder to optimize analytically.

Step 6 Reflection

X appears in dw because weights multiply inputs.
Bias gradient does not involve X because bias is independent of features.
If error becomes zero, gradients become zero and learning stops.

Step 8 Reflection

Loss should decrease over time.
If it increases, learning rate may be too high.
Learning rate controls speed; epochs control how long learning continues.

🔹 Final Concept Paragraph (Very Important)

This model is still the perceptron at its core because it performs the same linear computation: ŷ = Xw + b. The difference is that instead of using an activation function for classification, we directly output a numeric value. The loss function has changed from classification loss to Mean Squared Error. Learning still happens through gradient descent, where weights are adjusted to reduce error. The main difference from classification is that correctness is replaced by minimizing numeric error.
